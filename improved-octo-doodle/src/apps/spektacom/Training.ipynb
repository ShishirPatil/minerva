```python
import pandas as pd
import numpy as np
import tensorflow as tf
import keras as keras
import matplotlib.pyplot as plt
import os
#os.environ["CUDA_VISIBLE_DEVICES"]= "3"
```

```python
data_net = pd.read_csv("new_data.csv", header=None)


all_Y=np.zeros((130,1))
for i in range(0,130):
#     print(i+1, data_net[3][i*210])
    all_Y[i]=data_net[3][i*210]

coarse_Y=np.zeros((130,1))
for i in range(0,130):
    if(all_Y[i] in [21, 25, 29, 33, 37, 41]): 
        coarse_Y[i]=4
    if(all_Y[i] in [22,23,26,27,30,31,34,35,38,39,42,43]): 
        coarse_Y[i]=5
    if(all_Y[i] in [24,28,32,36,40,44]): 
        coarse_Y[i]=6
    if(all_Y[i] in [45,49,53,57]): 
        coarse_Y[i]=7
    if(all_Y[i] in [46,47,50,51,54,55,58,59]): 
        coarse_Y[i]=8
    if(all_Y[i] in [48,52,56,60]): 
        coarse_Y[i]=9
    if(all_Y[i] in [61,65,69]): 
        coarse_Y[i]=10
    if(all_Y[i] in [62,63,66,67,70,71]): 
        coarse_Y[i]=11
    if(all_Y[i] in [64,68,72]): 
        coarse_Y[i]=12


from numpy import reshape
X=np.zeros((130*150,1))
X0=data_net[:][0].values
X1=data_net[:][1].values
X2=data_net[:][2].values
X0=X0.reshape((130,210))
X0=X0[:,0:50]
X1=X1.reshape((130,210))
X1=X1[:,0:50]
X2=X2.reshape((130,210))
X2=X2[:,0:50]
X=np.concatenate((X0,X1,X2),axis=1)
print("X.shape:", X.shape)
Y=keras.utils.np_utils.to_categorical(coarse_Y-3)
from sklearn import preprocessing
X=preprocessing.scale(X)
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.utils import plot_model
```

```python
model = Sequential()
model.add(Dense(50,input_dim=150,activation='relu'))
model.add(Dense(10,input_dim=50,activation='sigmoid'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
model.fit(X_train, Y_train, epochs=50, batch_size=10,  verbose=2)

predictions = model.predict(X_test)
print(np.argmax(predictions, axis=1)+3)
print(np.argmax(Y_test,axis=1)+3)
```
